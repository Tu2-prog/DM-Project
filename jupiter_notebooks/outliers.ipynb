```python
import pandas as pd

# Function to clean up DataFrame by dropping specific columns
def clean_up_df(df, columns):
    return df.drop(columns=columns)

# Load the CSV file
df = pd.read_csv("../DBtrainrides_final_result.csv")

# Check if critical columns exist and contain non-NaN values
required_columns = ["arrival_delay_m", "ID_Base", "ID_Timestamp"]
if not all(col in df.columns for col in required_columns):
    raise ValueError(f"Missing required columns: {set(required_columns) - set(df.columns)}")
if df[required_columns].isnull().any().any():
    raise ValueError("Some required columns contain NaN values. Please clean the data.")

# Drop unnecessary columns
columns_to_drop = [
    "arrival_change", "departure_change", "departure_time", "cumulative_delay",
    "delay_gain", "station_progress", "planned_elapsed_time", "total_planned_time",
    "time_progress", "next_arrival_plan", "planned_travel_time_to_next_stop",
    "progress_ratio", "distance_from_origin", "distance_progress", "avg_city_delay"
]
df = clean_up_df(df, columns_to_drop)

# Calculate IQR for arrival_delay_m
Q1 = df["arrival_delay_m"].quantile(0.25)
Q3 = df["arrival_delay_m"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter outliers
filtered = df[(df["arrival_delay_m"] <= lower_bound) | (df["arrival_delay_m"] >= upper_bound)]
print("Original DataFrame:")
print(df)
print("Filtered Outliers DataFrame:")
print(filtered)

# Extract unique outlier IDs
outliers_ids = filtered[["ID_Base", "ID_Timestamp"]].drop_duplicates()

# Ensure outliers_ids is clean and usable
outliers_ids = outliers_ids.dropna()  # Remove rows with NaN values if any
outliers_ids_list = outliers_ids.apply(tuple, axis=1).tolist()  # Convert to list of tuples

# Filter the full dataset to exclude outliers
filtered_full_df = df.merge(
    outliers_ids, 
    on=["ID_Base", "ID_Timestamp"], 
    how="left", 
    indicator=True
).query('_merge == "left_only"').drop(columns=['_merge'])

print("Filtered Full DataFrame:")
print(filtered_full_df)
```
